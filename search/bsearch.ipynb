```python

"""
查找第一个值等于给定值的元素
"""
from typing import List


def bsearch(nums: List[int], target: int) -> int:
    low, high = 0, len(nums) - 1
    while low <= high:
        mid = low + (high - low) // 2
        if nums[mid] > target:
            high = mid - 1
        elif nums[mid] < target:
            low = mid + 1
        else:
            if mid == 0 or nums[mid - 1] != target:
                return mid
            else:
                high = mid - 1
    return -1


if __name__ == '__main__':
    array = [1, 2, 3, 4, 4, 4, 5, 6, 7, 8, 8, 8, 9, 9, 10]
    print(bsearch(array, 8))
```
